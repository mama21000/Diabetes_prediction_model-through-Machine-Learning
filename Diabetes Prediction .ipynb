### BUILD THE CLASSIFICATION ALGORITHMS

#### 1. LOGISTIC REGRESSION

```python
# MAKING A LOGISTIC REGRESSION MODEL USING TRAINING DATASET.

CLASSIFIERS=['Logisitc Regression','KNN','Naive Bayes','Support Vector Machines','Decision Tree','Random Forest','Xgboost']
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(C=5, penalty='l1', solver='liblinear')
lr.fit(X_train,Y_train)
```

```python
# HYPERARAMETER TUNING OF MODEL

from sklearn.model_selection import GridSearchCV
param_grid_lr=[{'penalty':['l1','l2','elasticnet'],
                'C' : [1,2,3,4,5,6,10,20],
                'max_iter':[100,200,300]
}]
grid_lr=GridSearchCV(lr,param_grid_lr)   # USING GRID SEARCH WITH 5 FOLDS CROSS VALIDATION
grid_lr.fit(X_train,Y_train)
```

```python
print(grid_lr.best_estimator_)
```

#### 2. KNeighborClassifier(KNN)

```python
# MAKING A KNN MODEL USING TRAINING DATASET.


from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(X_train,Y_train)
```

#### 3. Naive-Bayes Classifier

```python
# MAKING A NAUVE BAYES MODEL USING TRAINING DATASET.


from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(X_train,Y_train)
```

#### 4. Support Vector Machines(SVM)

```python
# MAKING A SVM MODEL USING TRAINING DATASET.


from sklearn.svm import SVC
svc=SVC(C=1, gamma=0.1)
svc.fit(X_train,Y_train)
```

```python
# HYPERARAMETER TUNING OF MODEL

from sklearn.model_selection import GridSearchCV
  
param_grid_svm = {'C': [0.1, 1, 10, 100, 1000], 
              'gamma': ['scale',1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']} 
  
grid_svm = GridSearchCV(svc, param_grid_svm, refit = True, verbose = 4)    # USING GRID SEARCH WITH 5 FOLDS CROSS VALIDATION
  
# fitting the model for grid search
grid_svm.fit(X_train, Y_train)
```

```python
print(grid_svm.best_params_)
print(grid_svm.best_estimator_)
```

#### 5. Decision Tree

```python
# MAKING A DECISION TREE MODEL USING TRAINING DATASET.


from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train,Y_train)
```

```python
 from sklearn import tree
text_representation = tree.export_text(dt)    # TEXT REPRESENTATION OF TREE
print(text_representation)
```

```python
# VISUALSISATION OF TREE

fig = plt.figure(figsize=(20,10))
tree_visual=tree.plot_tree(dt,
               filled=True)
```

### COST COMPLEXITY PRUNING

```python
# PRUNING THE TREE

path=dt.cost_complexity_pruning_path(X_train,Y_train)
ccp_alphas=path.ccp_alphas   # EXTRACTION ALPHA VALUES
ccp_alphas=ccp_alphas[:-1]   # REMOVING THE HIGHEST VALUE OF APLHA BECAUSE IT DENOTES ONLY ONE NODE

clf_dts=[]

for ccp_alpha in ccp_alphas:
    dt_prun_1=DecisionTreeClassifier(random_state=0,ccp_alpha=ccp_alpha)
    dt_prun_1.fit(X_train,Y_train)
    clf_dts.append(dt_prun_1)
    
```

```python
# PLOTTING SCORES OF TREES WITH DIFFERENT ALPHAS FOR TRAINING AND TESTING DATASET

train_scores=[dt_prun_1.score(X_train,Y_train) for dt_prun_1 in clf_dts]
test_scores=[dt_prun_1.score(X_test,Y_test) for dt_prun_1 in clf_dts]

fig,ax=plt.subplots()
ax.set_xlabel('alpha')
ax.set_ylabel('accuracy')
ax.set_title('accuracy vs value of alpha for training and testing dataset')
ax.plot(ccp_alphas,train_scores,marker='o',label='train')
ax.plot(ccp_alphas,test_scores,marker='o',label='test')
plt.show()
ccp_alphas
```

```python
#USING CROSS VALIDATION TO FIND THE BEST VALUE OF ALPHA WITH THE HIGHEST AVG. ACCURACY AMONG ALL THE FOLDS.

from sklearn.model_selection import cross_val_score

accuracy=[]

for ccp_alpha in ccp_alphas:

    dt_prun_2=DecisionTreeClassifier(random_state=0,ccp_alpha=ccp_alpha)
    scores=cross_val_score(dt_prun_2,X_train,Y_train,cv=5)
    
    accuracy.append(np.mean(scores))
    
    
    
plt.plot(ccp_alphas,accuracy,marker='o')
accuracy
    
    
```

##### SO WE CAN SEE THAT WE ARE GETTING A HIGHER ACCURACY FOR ALPHA = 0.00734199 AFTER CROSS VALIDATION  INSTEAD OF ALPHA =0.00683321

```python
#CREATING THE PRUNED TREE USING THE IDEAL VALUE OF ALPHA

ideal_ccp_alpha=0.00734199
dt=DecisionTreeClassifier(random_state=42,ccp_alpha=ideal_ccp_alpha)
dt_pruned=dt.fit(X_train,Y_train)
```

```python
#PLOTTONG THE PRUNED TREE

plt.figure(figsize=(15,7.5))
tree.plot_tree(dt_pruned,
               filled=True)
```

#### WE CAN SEE THE PRUNED TREE IS A BETTER TREE THAN BEFORE AS IT IS SMALLER AND IT WILL GIVE BETTER RESULT WITH THE TESTING DATASET
